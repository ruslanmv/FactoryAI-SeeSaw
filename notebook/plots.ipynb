```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV files
tokens1_path = 'tokens1.csv' ## Standard
tokens2_path = 'tokens2.csv' ## See-Saw
tokens1 = pd.read_csv(tokens1_path)
tokens2 = pd.read_csv(tokens2_path)

# Function for setting scientific plot styles
def set_plot_style():
    plt.rcParams.update({
        "font.size": 10,
        "axes.titlesize": 12,
        "axes.labelsize": 10,
        "xtick.labelsize": 9,
        "ytick.labelsize": 9,
        "legend.fontsize": 9,
        "figure.titlesize": 12
    })

# Set scientific plot style
set_plot_style()
```

```python

# 1. Comparison of Dependency Types
tokens1_summary = tokens1.groupby('Type')['Token Usage'].sum().reset_index()
tokens2_summary = tokens2.groupby('Type')['Token Usage'].sum().reset_index()

tokens1_summary['Method'] = 'Standard'
tokens2_summary['Method'] = 'See-Saw'

comparison = pd.concat([tokens1_summary, tokens2_summary], ignore_index=True)
comparison_pivot = comparison.pivot(index='Type', columns='Method', values='Token Usage')
```

```python
# Plot
fig, ax = plt.subplots(figsize=(6, 4))  # Adjust size for paper formatting
comparison_pivot.plot(kind='bar', ax=ax, color=['#1f77b4', '#ff7f0e'])
ax.set_title('Comparison of Dependency Types (Token Usage)')
ax.set_xlabel('Dependency Type')
ax.set_ylabel('Token Usage')
ax.legend(title='Method', loc='upper right')
plt.tight_layout()
plt.savefig("dependency_comparison.png", dpi=300)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV files
tokens1_path = 'tokens1.csv'
tokens2_path = 'tokens2.csv'
tokens1 = pd.read_csv(tokens1_path)
tokens2 = pd.read_csv(tokens2_path)

# Summarize data and exclude the "main" dependency type
tokens1_summary = tokens1.groupby('Type')['Token Usage'].sum().reset_index()
tokens2_summary = tokens2.groupby('Type')['Token Usage'].sum().reset_index()

# Filter out the "main" dependency type for the first plot
tokens1_filtered = tokens1_summary[tokens1_summary['Type'] != 'main'].copy()
tokens2_filtered = tokens2_summary[tokens2_summary['Type'] != 'main'].copy()

tokens1_filtered.loc[:, 'Method'] = 'Standard'
tokens2_filtered.loc[:, 'Method'] = 'See-Saw'

# Combine data for filtered dependencies comparison
comparison_filtered = pd.concat([tokens1_filtered, tokens2_filtered], ignore_index=True)
comparison_pivot_filtered = comparison_filtered.pivot(index='Type', columns='Method', values='Token Usage')

# Summarize data and filter for "main" dependency type for the second plot
tokens1_main = tokens1_summary[tokens1_summary['Type'] == 'main'].copy()
tokens2_main = tokens2_summary[tokens2_summary['Type'] == 'main'].copy()

tokens1_main.loc[:, 'Method'] = 'Standard'
tokens2_main.loc[:, 'Method'] = 'See-Saw'

# Combine data for main dependency comparison
comparison_main = pd.concat([tokens1_main, tokens2_main], ignore_index=True)
comparison_pivot_main = comparison_main.pivot(index='Type', columns='Method', values='Token Usage')

# Create a combined figure with two subplots
fig, axes = plt.subplots(1, 2, figsize=(16, 8))  # Wider and taller figure for professional layout

# Shared title
#fig.suptitle('Comparison of Token Usage in Code Generation', fontsize=16, weight='bold')

# Plot for filtered dependencies
comparison_pivot_filtered.plot(kind='bar', ax=axes[0], color=['#1f77b4', '#ff7f0e'])
axes[0].set_title('Dependencies Code Generation', fontsize=16)
axes[0].tick_params(axis='x', labelsize=14, rotation=0)  # Set x-axis labels horizontal
axes[0].tick_params(axis='y', labelsize=14)
axes[0].set_xlabel('')  # Remove x-axis label
axes[0].set_ylabel('Token Usage', fontsize=16)
axes[0].legend(title='Method', loc='upper right', fontsize=14)
axes[0].set_xticklabels(comparison_pivot_filtered.index, fontsize=16)  # Increase x-axis legend font size

# Plot for main dependency
comparison_pivot_main.plot(kind='bar', ax=axes[1], color=['#1f77b4', '#ff7f0e'])
axes[1].set_title('Main Code Generation', fontsize=16)
axes[1].tick_params(axis='x', labelsize=14, rotation=0)
axes[1].tick_params(axis='y', labelsize=14)
axes[1].set_xlabel('', fontsize=16)
axes[1].set_ylabel('Token Usage', fontsize=16)
axes[1].legend(title='Method', loc='upper right', fontsize=14)
axes[1].set_xticklabels(comparison_pivot_main.index, fontsize=16)  # Increase x-axis legend font size

# Adjust layout for clarity
plt.tight_layout(rect=[0, 0, 1, 0.93])  # Leave space for shared title
plt.savefig("dependency_comparison_new.png", dpi=300)
plt.show()
```
