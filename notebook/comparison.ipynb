```python
def check_path_columns(merged_df):
    """
    Adds a column 'Same Path' to the DataFrame to indicate whether 
    'path_standard' and 'path_seesaw' have the same value after removing prefixes 
    before the 'generated' directory. Also adds a 'normalized_path' column with 
    the matching normalized path or None if they do not match.

    Args:
        merged_df (pd.DataFrame): The merged DataFrame with columns 
                                  'path_standard' and 'path_seesaw'.

    Returns:
        pd.DataFrame: The updated DataFrame with the 'Same Path' and 'normalized_path' columns added.
    """
    def normalize_path(path):
        """
        Removes the prefix before 'generated' in the given path.

        Args:
            path (str): The file path to normalize.

        Returns:
            str: The normalized path starting from 'generated'.
        """
        if "generated" in path:
            return path.split("generated", 1)[-1]
        return path  # Return the path unchanged if 'generated' is not found

    # Normalize the paths
    merged_df['normalized_path_standard'] = merged_df['path_standard'].apply(normalize_path)
    merged_df['normalized_path_seesaw'] = merged_df['path_seesaw'].apply(normalize_path)

    # Compare the normalized paths
    merged_df['Same Path'] = (
        merged_df['normalized_path_standard'] == merged_df['normalized_path_seesaw']
    )

    # Add 'normalized_path' column
    merged_df['normalized_path'] = merged_df.apply(
        lambda row: row['normalized_path_standard'] if row['Same Path'] else None,
        axis=1
    )

    # Optionally drop intermediate columns if no longer needed
    merged_df.drop(columns=['normalized_path_standard', 'normalized_path_seesaw'], inplace=True)

    return merged_df
```

```python
def compute_differences(df, column1, column2, new_column_name='differences'):
    """
    Adds a new column to the DataFrame that computes the differences 
    between two text columns, with clear delimiters to separate the first 
    and second code differences.

    Args:
        df (pd.DataFrame): The input DataFrame containing the text columns.
        column1 (str): The name of the first column for comparison.
        column2 (str): The name of the second column for comparison.
        new_column_name (str): The name of the new column to store the differences.

    Returns:
        pd.DataFrame: The DataFrame with the added 'differences' column.
    """
    def calculate_diff(text1, text2):
        """
        Calculate the line-by-line differences between two strings.

        Args:
            text1 (str): First text.
            text2 (str): Second text.

        Returns:
            str: A string showing the differences with delimiters.
        """
        diff = list(ndiff(str(text1).splitlines(), str(text2).splitlines()))
        
        # Add delimiters
        delimiter_start = "----- START OF FIRST CODE -----\n"
        delimiter_end = "\n----- START OF SECOND CODE -----\n"
        
        # Separate the first and second code differences
        first_code_diff = "\n".join(line for line in diff if line.startswith("- ") or line.startswith("  "))
        second_code_diff = "\n".join(line for line in diff if line.startswith("+ ") or line.startswith("  "))
        
        return delimiter_start + first_code_diff + delimiter_end + second_code_diff

    # Apply the diff function row-wise
    df[new_column_name] = df.apply(
        lambda row: calculate_diff(row[column1], row[column2]), axis=1
    )
    
    return df
```
